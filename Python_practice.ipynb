```python
#Create three classes: LivingBeing, Animal, and Bird where:
#LivingBeing has is_alive = True.
#Animal inherits from LivingBeing.
#Bird inherits from Animal and has a fly() method.
#Instantiate Bird and demonstrate inherited attributes/methods
class LivingBeing:
  def __init__(self):
    self.is_alive = True
class Animal(LivingBeing):
  def __init__(self):
    super(). __init__()
class Bird(Animal):
  def __init__(self, breed):
    super(). __init__()
    self.breed = breed

  def fly(self):
    return f"The {self.breed} is flying!"

Humming = Bird("Humming")

print(Humming.is_alive)
print(Humming.fly())
```

```python
#Create an iterator class EvenNumbers that takes a limit and iterates over even numbers only. Use it in a for loop.
class EvenNumbers:
  def __iter__(self):
    self.a = 0
    return self
  def __next__(self):
    for i in range(self.a, self.a + 2, 2):
            if i >= 10:
                raise StopIteration
            self.a = i + 2
            return i

evenclass= EvenNumbers()
for number in evenclass:
    print(f"The even {number} number")

```

```python
#Create classes Rectangle, Circle, and Triangle, each with an area() method.
#Store instances of these classes in a list.
#Write code that calculates the total area without checking object types
import math

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

class Triangle:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


shapes = [
    Rectangle(10, 20),
    Circle(7),
    Triangle(5, 12)
]

total_area = sum(shape.area() for shape in shapes)
print(f"Total area: {total_area}")



```

```python
#Library Management System
#1 Book class
#Oop and inheritance
class Book:
  borrowed_books = 0

  def __init__(self, title, author, year):
    self.title = title
    self.author = author
    self.year = year
    self.available = True
  def get_description(self):
    return f"{self.title} by {self.author} ({self.year})"

class Ebook(Book):
  def __init__(self,title,author, year, file_size, file_format):
     super().__init__(title, author, year)
     self.file_size = file_size
     self.file_format = file_format

  def get_description(self):
    return f"[EBook] {super().get_description()} - {self.file_size}MB, {self.file_format}"

class PrintedBook(Book):
  def __init__(self, title, author, year, weight, dimensions):
    super().__init__(title, author, year)
    self.weight = weight
    self.dimension = dimensions

  def get_description(self):
    return f"[EBook] {super().get_description()} - {self.weight}MB, {self.dimension}"

class Member:
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self._borrowed_books = []

    def borrow_book(self, book):
        raise NotImplementedError("This method should be overridden.")

    def get_borrowed_books(self):
        return [book.title for book in self._borrowed_books]

    def __iter__(self):
        return iter(self._borrowed_books)

class StudentMember(Member):
    def borrow_book(self, book):
        if len(self._borrowed_books) >= 3:
            print(f"{self.name} (Student) can't borrow more than 3 books.")
            return
        if book.available:
            self._borrowed_books.append(book)
            book.available = False
            Book.borrowed_books += 1
            print(f"{self.name} borrowed '{book.title}'")
        else:
            print(f"'{book.title}' is not available.")

class TeacherMember(Member):
    def borrow_book(self, book):
        if len(self._borrowed_books) >= 6:
            print(f"{self.name} (Teacher) can't borrow more than 6 books.")
            return
        if book.available:
            self._borrowed_books.append(book)
            book.available = False
            Book.borrowed_books += 1
            print(f"{self.name} borrowed '{book.title}'")
        else:
            print(f"'{book.title}' is not available.")

# 3. Library class with iterators and private attributes
class Library:
    def __init__(self, name):
        self.__library_name = name  # Private attribute
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def search_by_title(self, title):
        return [book.get_description() for book in self.books if title.lower() in book.title.lower()]

    def search_by_author(self, author):
        return [book.get_description() for book in self.books if author.lower() in book.author.lower()]

    def __iter__(self):
        return iter(self.books)

    def get_library_name(self):
        return self.__library_name

# 4. Simple usage example
library = Library("City Library")

# Add books
library.add_book(Ebook("Python Basics", "Alice", 2022, "130Mb", "PDF"))
library.add_book(PrintedBook("Data Science", "Nancy", 2018, "1.5", "22x15 cm"))
library.add_book(PrintedBook("AI Future", "James", 2021, "2", "25x18 cm"))

# Add members
student = StudentMember("John", 101)
teacher = TeacherMember("Ms. Smith", 201)

# Borrow books
for book in library:
    if book.title == "Python Basics":
        student.borrow_book(book)
    elif book.title == "Data Science":
        teacher.borrow_book(book)

# Search
print("\nSearch Results for author 'Alice':")
for desc in library.search_by_author("Alice"):
    print(desc)

# Showing borrowed books
print("\nBooks borrowed by student:", student.get_borrowed_books())
print("Books borrowed by teacher:", teacher.get_borrowed_books())

# Total borrowed
print(f"\nTotal books borrowed by all members: {Book.borrowed_books}")
```
